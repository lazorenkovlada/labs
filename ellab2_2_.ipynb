```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=[20, 50, 100, 200, 500, 1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 20000, 40000, 60000, 100000, 200000, 500000, 1000000, 2000000]
Ф=[90, 90, 90, 90, 90, 81, 72, 70, 57, 52, 45, 40, 36, 27, 14, 9, 6, 5, 2, 1, 0]
```

```python
plt.figure(figsize=(10, 6))
plt.semilogx(f, Ф, 'bo-', linewidth=2, markersize=6, label='Експеримент') 
plt.title('Фазово-частотна характеристика (АЧХ)')
plt.xlabel('Частота f, Гц (лог. шкала)')
plt.ylabel('Ф, град')
plt.grid(True, which="both", ls="-", alpha=0.5) # Сітка в клітинку
plt.legend() 

plt.show()
```
