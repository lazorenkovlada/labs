```python
plt.figure(figsize=(10, 6))
plt.plot(u, I, 'bo-', linewidth=2, markersize=6, label='Експеримент') 
plt.title('ВАХ')
plt.xlabel('U, V')
plt.ylabel('I, A')
plt.grid(True, which="both", ls="-", alpha=0.5) # Сітка в клітинку
plt.legend() 

plt.show()
```
