```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f=[20, 50, 100, 200, 500, 1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 20000, 40000, 60000, 100000, 200000, 500000, 1000000, 2000000]
k=[0.0021, 0.00496, 0.01, 0.02, 0.0497, 0.1, 0.197, 0.373, 0.519, 0.622, 0.718, 0.766, 0.836, 0.892, 0.968, 0.994, 1, 1, 1, 1, 1]
```

```python
K_db=20*np.log10(k)
```

```python
plt.figure(figsize=(10, 6))
plt.semilogx(f, K_db, 'bo-', linewidth=2, markersize=6, label='Експеримент') 
plt.title('Амплітудно-частотна характеристика (АЧХ)')
plt.xlabel('Частота f, Гц (лог. шкала)')
plt.ylabel('Коефіцієнт передачі K, дБ')
plt.grid(True, which="both", ls="-", alpha=0.5) # Сітка в клітинку
plt.legend() 

plt.show()
```
